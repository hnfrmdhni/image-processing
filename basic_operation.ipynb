```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# mengaktifkan OpenCL pada OpenCV
cv2.ocl.setUseOpenCL(True)

# menampilkan informasi OpenCL yang tersedia
print(cv2.ocl.haveOpenCL())
print(cv2.ocl.useOpenCL())

# daftar nama file gambar yang akan diproses
file_names = ['gambar1.jpg', 'gambar2.jpg', 'gambar3.jpg', 'gambar4.jpg', 'gambar5.jpg','gambar6.jpg', 'gambar7.jpg', 'gambar8.jpg', 'gambar9.jpg', 'gambar10.jpg']

# operasi dasar pengolaan citra
for name in file_names:
    # membaca gambar
    file_path = f'resources/{name}'
    img = cv2.imread(file_path)

    # thresholding
    thresh_value = 100
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    _, thresh = cv2.threshold(gray, thresh_value, 255, cv2.THRESH_BINARY)

    # grayscale
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # image brightening
    bright = cv2.addWeighted(img, 1.5, np.zeros(img.shape, img.dtype), 0, 0)

    # operasi aritmatik
    img1 = cv2.imread('resources/gambar1.jpg')
    img2 = cv2.imread('resources/gambar2.jpg')
    add = cv2.add(img1, img2)
    subtract = cv2.subtract(img1, img2)
    multiply = cv2.multiply(img1, img2)
    divide = cv2.divide(img1, img2)

    # operasi boolean
    img1 = cv2.imread('resources/gambar1.jpg')
    img2 = cv2.imread('resources/gambar2.jpg')
    and_op = cv2.bitwise_and(img1, img2)
    or_op = cv2.bitwise_or(img1, img2)
    xor_op = cv2.bitwise_xor(img1, img2)
    not_op = cv2.bitwise_not(img1)

    # operasi geometri
    rows, cols = img.shape[:2]
    M = cv2.getRotationMatrix2D((cols/2,rows/2), 45, 1)
    rotate = cv2.warpAffine(img, M, (cols, rows))
    flip_hor = cv2.flip(img, 1)
    flip_ver = cv2.flip(img, 0)

    # menampilkan hasil
    fig, axs = plt.subplots(nrows=4, ncols=4, figsize=(12, 12))
    axs[0, 0].imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    axs[0, 0].set_title('Original')
    axs[0, 1].imshow(thresh, cmap='gray')
    axs[0, 1].set_title('Thresholding')
    axs[0, 2].imshow(gray, cmap='gray')
    axs[0, 2].set_title('Grayscale')
    axs[0, 3].imshow(cv2.cvtColor(bright, cv2.COLOR_BGR2RGB))
    axs[0, 3].set_title('Brightening')
    axs[1, 0].imshow(cv2.cvtColor(add, cv2.COLOR_BGR2RGB))
    axs[1, 0].set_title('Addition')
    axs[1, 1].imshow(cv2.cvtColor(subtract, cv2.COLOR_BGR2RGB))
    axs[1, 1].set_title('Subtraction')
    axs[1, 2].imshow(cv2.cvtColor(multiply, cv2.COLOR_BGR2RGB))
    axs[1, 2].set_title('Multiplication')
    axs[1, 3].imshow(cv2.cvtColor(divide, cv2.COLOR_BGR2RGB))
    axs[1, 3].set_title('Division')
    axs[2, 0].imshow(cv2.cvtColor(and_op, cv2.COLOR_BGR2GRAY))
    axs[2, 0].set_title('AND')
    axs[2, 1].imshow(cv2.cvtColor(or_op, cv2.COLOR_BGR2RGB))
    axs[2, 1].set_title('OR')
    axs[2, 2].imshow(cv2.cvtColor(xor_op, cv2.COLOR_BGR2RGB))
    axs[2, 2].set_title('XOR')
    axs[2, 3].imshow(cv2.cvtColor(not_op, cv2.COLOR_BGR2RGB))
    axs[2, 3].set_title('NOT')
    axs[3, 0].imshow(cv2.cvtColor(rotate, cv2.COLOR_BGR2RGB))
    axs[3, 0].set_title('Rotation')
    axs[3, 1].imshow(cv2.cvtColor(flip_hor, cv2.COLOR_BGR2RGB))
    axs[3, 1].set_title('Horizontal Flip')
    axs[3, 2].imshow(cv2.cvtColor(flip_ver, cv2.COLOR_BGR2RGB))
    axs[3, 2].set_title('Vertical Flip')
    axs[3, 3].axis('off')

    # menampilkan plot
    plt.show() 
```
